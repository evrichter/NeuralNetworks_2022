### 1.1

The Jacobian matrix is a coordinate-based matrix representation of the derivative of a vector-valued or a multivariable function when the derivative of that function exists.
    
$J_{f1, f2}(x,y) = \begin{bmatrix} \frac{\delta f_{1}} {\delta x} & \frac{\delta f_{1}} {\delta y}\\ \frac{\delta f_{2}} {\delta x} &  \frac{\delta f_{2}} {\delta y} \\  \end{bmatrix} = \begin{bmatrix} 10xy^{2} + 10x^{4}y^{3} - 32x^{3}y^{3} & 10x^{2}y + 6x^{5}y^{2} - 24x^{4}y^{2}\\ 3x^{2}y^{2} + 6xy +e^{2y} & 2x^{3}y + 3x^{2} + 2xe^{2y}\\  \end{bmatrix}$
    
The Hessian matrix of a multivariable function (f(x,y,z,...)) is a square matrix of second-order partial derivatives of a scalar-valued function, or scalar field. 
    
$H_{f_{1}}(x,y) = \begin{bmatrix} \frac{\delta^{2}f} {\delta x^{2}} & \frac{\delta^{2}f} {\delta x \delta y}\\ \frac{\delta^{2}f} {\delta y \delta x} & \frac{\delta^{2}f} {\delta y^{2}} \\  \end{bmatrix} = \begin{bmatrix} 10y^{2} + 40x^{3}y^{3} - 96x^{2}y^{3} & 20xy + 30x^{4}y^{2} - 96x^{3}y^{2}\\ 20xy + 30x^{4}y^{2} - 96x^{3}y^{2} & 10x^{2} + 12x^{5}y - 48x^{4}y \\  \end{bmatrix} $
    
$H_{f_{2}}(x,y) = \begin{bmatrix} \frac{\delta^{2}f} {\delta x^{2}} & \frac{\delta^{2}f} {\delta x \delta y}\\ \frac{\delta^{2}f} {\delta y \delta x} & \frac{\delta^{2}f} {\delta y^{2}} \\  \end{bmatrix} = \begin{bmatrix} 6xy^{2} + 6y & 6x^{2}y + 6x + 2e^{2y}\\ 6x^{2}y + 6x + 2e^{2y} & 2x^{3} + 4xe^{2y} \\  \end{bmatrix} $

### 1.2

The Hessian matrix is expensive to compute, but it is useful because it can help to determine local maxima, local minima and saddle points.

1. If the Hessian matrix is positive defnite (<=> all eigenvalues are positive), this indicates a minimum. 

2. If the Hessian matrix is negative definite (<=> all eigenvalues are negative), this indicates a maximum. 

3. If the Hessian matrix is indefinite (<=> has positive and negative eigenvalues), this indicates a saddle point.

4. If the determinant equals zero the Hessian matrix is undeterminable without further information/visualization.

### 2.4
Gradient descent tries to find a local minimum using only the first derivative of a function f, based on which it iteratively searches for the path to the minimum starting from a certain guess by following the steepest descent step by step. Since it only considers the first derivative, it can take a long time to converge, especially in cases where a landscape has many local curves. 

Newton's method on the other hand is a second order method as it assumes that loss is twice differentiable and uses the approximation with the Hessian matrix (consisting of second derivatives). More precisely, the inverse of the Hessian matrix is multiplied by the gradient of the function, which greatly reduces the number of steps needed to find the minimum, i.e. convergence occurs faster, at the cost of being computationally very expensive. 
Precisely because of this reason Newton's method is rarely used in machine learning despite its apparent superiority over gradient descent. This is because Newton's method maximises a function knowing its second derivative. Ideally, if the second derivative is very simple to calculate or given, this is faster, but in practice the second derivative is often a complex expression and solving it requires a lot of computational effort. Hence, it is usually computationally much more feasible to use gradient descent.

In summary, both are optimisation methods for finding critical points (minimum, maximum and saddle points), but Newton's method has stronger constraints on the differentiability of the function than gradient descent. Therefore, when Newton's method converges, it does so more quickly than gradient descent. On the other hand, gradient descent converges more often and is used more often since using Newton's method it is computationally expensive to find the critical points for most functions.
