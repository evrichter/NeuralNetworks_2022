1. Transpose of a matrix  = matrix that is obtained after reversing its rows to columns or columns to rows.

* Example: $A = \begin{bmatrix}3 & 4 & 7\\7 & 3 & 0\end{bmatrix} A^{T} = \begin{bmatrix}3 & 7\\4 & 3 \\1 & \frac{3}{5}\end{bmatrix} $ 


2. Identity matrix = square matrix in which all the elements of principal diagonals are one, and all other elements are zeros.

* Example for 3X3 matrix: I = $\begin{bmatrix}1 & 0 & 0\\0 & 1 & 0\\ 0 & 0 & 1\end{bmatrix}$

3. Inverse of a matrix = another matrix, which on multiplying with the given matrix gives the multiplicative identity. The matrix inverse can only be found for square matrices, whose number of rows and columns are equal. 

* Example: $A X A^{T} = I: \begin{bmatrix} -3 & 1\\5 & 0\end{bmatrix} X \begin{bmatrix}0 &\frac{1}{5} \\1 & \frac{3}{5}\end{bmatrix} = \begin{bmatrix} 1 & 0\\0 & 1\end{bmatrix}$ 

4. Symmetric matrix = a matrix is symmetric if it is equal to its transpose, i.e. if swapping indices doesn't change its components.

* Example: $A = \begin{bmatrix}3 & 4 & 7\\7 & 3 & 0\end{bmatrix} A^{T} = \begin{bmatrix}3 & 4 & 7\\7 & 3 & 0\end{bmatrix}$ 

5. Orthogonal matrix = when the product of a square matrix and its transpose gives an identity matrix, then the square matrix is called orthogonal matrix.

* Example: $A X A^{T} = I: \begin{bmatrix} \frac{1}{3} & \frac{2}{3} & -\frac{2}{3} \\ -\frac{2}{3} & \frac{2}{3} & \frac{1}{3} \\ \frac{2}{3} & \frac{1}{3} & \frac{2}{3}\end{bmatrix} X \begin{bmatrix} \frac{1}{3} & -\frac{2}{3} & \frac{2}{3} \\ \frac{2}{3} & \frac{2}{3} & \frac{1}{3} \\ -\frac{2}{3} & \frac{1}{3} & \frac{2}{3}\end{bmatrix} = 
\begin{bmatrix} 1 & 0 & 0\\0 & 1 & 0\\ 0 & 0 & 1\end{bmatrix}$

```python
# TODO
# NUMPY

rand_int = np.random.randint(21,26,30)  #create numpy array with random integers
print(rand_int)

mean = np.mean(rand_int) #calculate mean using Numpy
print(mean)

def my_mean(numbers): #calculate mean with own function
    sum_numbers = 0
    for number in numbers:
        sum_numbers = sum_numbers + number       

    mean = sum_numbers / len(numbers)
    return mean

print(my_mean(rand_int)) #the mean calculated using numpy equals the mean calculated using our function

sd = np.std(rand_int) #calculate sd using Numpy
print(sd)

def my_sd(numbers): #calculate sd with own function
    n = len(numbers)
    mean = sum(numbers) / n
    variance = sum((number - mean)**2 for number in numbers) / n
    std_dev = variance ** 0.5
    return std_dev

print(my_sd(rand_int))  #the sd calculated using numpy equals the sd calculated using our function
```
